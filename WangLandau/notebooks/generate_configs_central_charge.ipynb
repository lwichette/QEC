```python
import pandas as pd
from tabulate import tabulate
import numpy as np
```

### Configs for prerun and main run

```python
alpha = 0.8

beta = [1e-08]

walker_wl = 4

overlap = 0.25

seed_hist = 1

seed_run = 1

num_interactions = 100

replica_exchange_offset = 50

boundary_type = 0

intervals_wl = 167

iterations = 1000

time_limit = 10800

num_loops = 1000

num_walker_prerun = 512
```

```python
row = []
array_task = 999
```

```python
for b in beta:
    for p in [0]:
        for x in [4, 5, 6, 7, 8]:
            for y in [128, 256, 512]:
                for error in ["I"]:
                    intervals_wl = y+1
                    num_intervals_prerun = 2*x*y
                    row.append(
                        [
                            array_task, alpha, b, walker_wl, overlap, seed_hist, seed_run, num_interactions, replica_exchange_offset,
                            boundary_type, intervals_wl, iterations, time_limit, num_loops, num_walker_prerun, num_intervals_prerun, p, x, y, error
                        ]
                    )
                    array_task += 1
                    seed_run += 1
```

```python
df = pd.DataFrame(row, columns=["ArrayTaskID", "alpha", "beta", "walker_wl", "overlap", "seed_hist", "seed_run",
                                "num_interactions", "replica_exchange_offset", "boundary_type", "intervals_wl",
                                "iterations", "time_limit", "num_loops", "num_walker_prerun", "num_intervals_prerun", "probability", "x", "y", "error_type"])
```
