```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
os.chdir('results')
```

```python
os.listdir('low_temperature')
```

```python
L_12_high = np.loadtxt('comparison_old_results/L_12_ns_200_ne_200_ni_1000_nw_100.txt')
L_36_high = np.loadtxt('comparison_old_results/L_36_ns_200_ne_200_ni_5000_nw_100.txt')
```

```python
L_12_low = np.loadtxt('low_temperature/L_12_ns_200_ne_200_ni_100000_nw_100.txt')
L_36_low = np.loadtxt('low_temperature/L_36_ns_200_ne_200_ni_50000_nw_100.txt')
```

```python
plt.plot(L_36_high[:,1], L_36_high[:,0], marker = 'o', label = "L=36_high")
plt.plot(L_12_high[:,1], L_12_high[:,0], marker = 'o', label = "L=12_high", color = )

plt.plot(L_36_low[:,1], L_36_low[:,0], marker = 'o', label = "L=36_low")
plt.plot(L_12_low[:,1], L_12_low[:,0], marker = 'o', label = "L=12_low")

plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.axhline(1, color = 'orange')
plt.legend()
```

```python
res_dict = {}
```

```python
for l in os.listdir():
    res_dict[int(l.split('L_')[1].split('_')[0])] = np.loadtxt(l)
```

```python
for l in sorted(res_dict.keys()):
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = "L="+str(l))
    
plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.axhline(1, color = 'orange')
plt.legend()
```
