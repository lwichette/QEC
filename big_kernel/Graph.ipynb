```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
os.chdir('results/comparison_old_results')
```

```python
res_dict = {}
```

```python
for l in os.listdir():
    res_dict[int(l.split('L_')[1].split('_')[0])] = np.loadtxt(l)
```

```python
for l in sorted(res_dict.keys()):
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = "L="+str(l))
    
plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.axhline(1, color = 'orange')
plt.legend()
```
