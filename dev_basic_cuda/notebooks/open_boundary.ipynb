```python
os.chdir('./dev_cluster')
os.getcwd()

res_dict={}

for l in os.listdir():
    if "up_1" in l:
        size = int(l.split('_')[1])
        res_dict[size] = np.loadtxt(l)

res_dict = {k: res_dict[k] for k in sorted(res_dict)}

plt.figure(figsize=(10,6))

for l in res_dict.keys():
    plt.plot(res_dict[l][:,1], 1/(2*l*m.sin(m.pi/l))*res_dict[l][:,0], marker = 'o', label = int(l))


# plt.xlim([0.95, 1.3])
# plt.ylim([0, 1.75])
plt.xlabel("T")
plt.ylabel("$ \\xi $ / L ")
plt.legend()
os.chdir('..')
```
