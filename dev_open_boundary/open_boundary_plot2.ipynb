```python
import os
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python

os.getcwd()
```

```python
os.chdir('./comparison_warmups')
os.getcwd()

res_dict={}
for l in os.listdir():
    if "lo_" in l:
        tag = int(l.split('_')[11])
        res_dict[tag] = np.loadtxt(l)
    else:
        tag = int(l.split('_')[9])
        res_dict[tag] = np.loadtxt(l)

res_dict = {k: res_dict[k] for k in sorted(res_dict)}

plt.figure(figsize=(10,6))

for l in res_dict.keys():
    plt.plot(res_dict[l][:,1], 1/(64*m.sin(m.pi/64))*res_dict[l][:,0], marker = 'o', label = "wu_"+str(l))


# plt.xlim([1.0, 2.0])
# plt.ylim([0, 1.75])
plt.xlabel("T")
plt.ylabel("$ \\zeta $ / L ")
plt.legend()
os.chdir('..')
```

```python
os.chdir('./dev_cluster')
os.getcwd()

res_dict={}
for l in os.listdir():
    size = int(l.split('_')[1])
    res_dict[(l,size)] = np.loadtxt(l)

res_dict = {k: res_dict[k] for k in sorted(res_dict)}

plt.figure(figsize=(10,6))

for (l,size) in res_dict.keys():
    if size == 64:
        plt.plot(res_dict[(l,size)][:,1], 1/(size*m.sin(m.pi/size))*res_dict[(l,size)][:,0], marker = 'o', label = str(l))


# plt.xlim([1.0, 2.0])
# plt.ylim([0, 1.75])
plt.xlabel("T")
plt.ylabel("$ \\zeta $ / L ")
plt.legend()
os.chdir('..')
```
