```python
os.chdir('./p10_estimate')
os.getcwd()

res_dict={}
for l in os.listdir():
    size = int(l.split('_')[1])
    res_dict[(l, size)] = np.loadtxt(l)

res_dict = {k: res_dict[k] for k in sorted(res_dict)}

plt.figure(figsize=(10,6))

for (l, size) in res_dict.keys():
    if not "lo" in l and not "write_lattice" in l:
        plt.plot(res_dict[(l, size)][:,1], 1/(2*size*m.sin(m.pi/size))*res_dict[(l, size)][:,0], marker = 'o', label = str(l))

    elif "write_lattice" in l:
        plt.plot(res_dict[(l, size)][:,1], 1/(size)*res_dict[(l, size)][:,0], marker = 'o', label = str(l))

    else:
        plt.plot(res_dict[(l, size)][:,1], 1/(2*size*m.sin(m.pi/size))*res_dict[(l, size)][:,0], marker = 'o', label = str(l))

# plt.xlim([1.0, 2.0])
# plt.ylim([0, 1.75])
plt.xlabel("1/T")
plt.ylabel("$ \\chi $ / L ")
plt.legend()
os.chdir('..')
os.getcwd()
```

```python
os.chdir('/home/dfki.uni-bremen.de/lpwichette/Desktop/Repos/qec/dev_open_boundary/results/open_boundary/dev_cluster')
os.getcwd()

for l in os.listdir():
    size = int(l.split('_')[1])
    print(l)
    with open(l, 'r') as f:
        data = f.readlines()
    rescaled_data = []
    for line in data:
        values = line.strip().split()
        if len(values) >= 2 and not 'write_lattice' in l:
            print(values)
            rescaled_value = 1/(2*m.sin(m.pi/size))*float(values[0])
            new_line = f"{rescaled_value} {' '.join(values[1:])}\n"
            rescaled_data.append(new_line)

    output_file = os.path.join("../dev_cluster_rescaled/", f"{l}")
    with open(output_file, 'w') as f:
        f.writelines(rescaled_data)

os.chdir('..')
```

```python
os.chdir('/home/dfki.uni-bremen.de/lpwichette/Desktop/Repos/qec/dev_open_boundary/results/open_boundary/p10_estimate')
os.getcwd()

for l in os.listdir():
    size = int(l.split('_')[1])
    print(l)
    with open(l, 'r') as f:
        data = f.readlines()
    rescaled_data = []
    for line in data:
        values = line.strip().split()

        if len(values) >= 2 and not "lo" in l and not "write_lattice" in l:
            rescaled_value = 1/(2*m.sin(m.pi/size))*float(values[0])
            new_line = f"{rescaled_value} {' '.join(values[1:])}\n"
            rescaled_data.append(new_line)

        elif len(values) >= 2 and "write_lattice" in l:
            rescaled_value = float(values[0])
            new_line = f"{rescaled_value} {' '.join(values[1:])}\n"
            rescaled_data.append(new_line)

        elif len(values) >= 2:
            rescaled_value = 1/(2*m.sin(m.pi/size))*float(values[0])
            new_line = f"{rescaled_value} {' '.join(values[1:])}\n"
            rescaled_data.append(new_line)

    output_file = os.path.join("../p10_estimate_rescaled/", f"{l}")
    with open(output_file, 'w') as f:
        f.writelines(rescaled_data)

os.chdir('..')
```
