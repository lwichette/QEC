```python
import matplotlib.pyplot as plt
import numpy as np

import os
```

```python
L = [12, 14]

res_dict = {}
```

```python
folder_name = "low_temp/"

d = os.listdir("results/"+folder_name)
```

```python
for l in L:
    res = [i for i in d if "psi_L_"+str(l) in i]
    res_dict[l] = np.loadtxt("results/"+folder_name+res[0])
    
for l in L:
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = "L="+str(l))
    
plt.xlabel("Probability p")
plt.ylabel("$ \\xi $ / L")
plt.legend()
```
