```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
open_boundary = True
```

```python
if open_boundary:
    os.chdir('open_boundary/')
else:
    os.chdir('periodic_boundary/')
```

```python
probs = [float(p) for p in os.listdir() if '.' in p]
```

```python
probs.sort()
```

```python
for prob in probs:
    os.chdir(str(prob))
    res_dict = {}

    for l in os.listdir():
        if "up" in l:
            size = int(l.split('_')[1])
            res_dict[size] = np.loadtxt(l)

    res_dict = {k: res_dict[k] for k in sorted(res_dict)}

    plt.figure(figsize=(10,6))

    for l in res_dict.keys():
        plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = int(l))
    plt.axvline(1.24, c='r')
    plt.xlabel("Temperature T")
    plt.ylabel("$ \\xi $ / L")
    plt.title(f"prob: {prob}, updates: (64, 500k), (128, 500k), (256,500k)")
    plt.legend() 
    os.chdir('..')
```
