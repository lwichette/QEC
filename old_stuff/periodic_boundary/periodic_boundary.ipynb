```python
L = {12, 14, 18, 24, 28, 36}
num_iters = {1000, 5000}
```

```python
df_concat = []
```

```python
for l in L:
    if l in {12,14,18}:
        L_run = np.loadtxt(f'reps/L_{l}_p_0.060000_ns_200_ne_200_ni_{1000}_nw_100.txt')
    else: 
        L_run = np.loadtxt(f'reps/L_{l}_p_0.060000_ns_200_ne_200_ni_{5000}_nw_100.txt')
    
    df_run = pd.DataFrame(L_run, columns = ["Correlation Length", "Temperature"])
    
    df_run['Lattice Size'] = l
    
    df_concat.append(df_run)
```

```python
df = pd.concat(df_concat)
```

```python
sns.lineplot(x="Temperature", y="Correlation Length", data=df, hue = "Lattice Size", palette = "tab10")
```
