```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir('results/')
```

```python
start_temp = 0.1091
step = 0.00001
```

```python
p = [np.round(start_temp + i*step,6) for i in range(12)]
```

```python
for prob in p:
    os.chdir(str(prob))
    
    res_dict = {}
    
    for l in os.listdir():
        size = int(l.split('_')[1])
        res_dict[size] = np.loadtxt(l)
    
    res_dict = {k: res_dict[k] for k in sorted(res_dict)}
    
    plt.figure(figsize=(10,6))

    for l in res_dict.keys():
        plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = int(l))

    plt.xlabel("Temperature T")
    plt.ylabel("$ \\xi $ / L")
    plt.title(f"prob: {prob}, updates: (64, 200k), (128, 300k), (256,400k), (512, 500k)")
    plt.legend()
    
    os.chdir('..')
```
