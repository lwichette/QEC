```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir('results/')
```

```python
p = [0.09 + i*0.0025 for i in range(4)]
```

```python
p = p + [0.1]
```

```python
os.chdir('0.1')
```

```python
res_dict = {}

for l in os.listdir():
    if '200000' in l:
        size = int(l.split('_')[1]) 
        res_dict[size] = np.loadtxt(l)

res_dict = {k: res_dict[k] for k in sorted(res_dict)}

plt.figure(figsize=(10,6))

for l in res_dict.keys():
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = int(l))

plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.title(f"prob: 0.1")
plt.legend()
```

```python
for prob in p:
    os.chdir(str(prob))
    
    res_dict = {}
    
    for l in os.listdir():
        size = int(l.split('_')[1])
        
        if size in [12, 36, 64]:
            if prob == 0.09:
                if size == 12:
                    np.loadtxt('L_12_p_0.090000_ns_10_ne_500_ni_40000_nw_100.txt')
            res_dict[size] = np.loadtxt(l)
    
    res_dict = {k: res_dict[k] for k in sorted(res_dict)}
    
    plt.figure(figsize=(10,6))

    for l in res_dict.keys():
        plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = int(l))

    plt.xlabel("Temperature T")
    plt.ylabel("$ \\xi $ / L")
    plt.title(f"prob: {prob}")
    plt.legend()
    
    os.chdir('..')
```
