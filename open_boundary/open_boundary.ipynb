```python
res_dict = {}
```

```python
for l in os.listdir():
    size = l.split('_')[1]
    res_dict[size] = np.loadtxt(l)
```

```python
for l in res_dict.keys():
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = "L="+str(l))
    
plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.legend()
```
