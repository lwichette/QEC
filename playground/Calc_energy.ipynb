```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = [28, 36, 50]
```

```python
res_dict = {}
```

```python
for l in L:
    res_dict[l] = np.loadtxt("results/low_temperature/psi_L_" +str(l)+ ".txt")
```

```python
for l in L:
    plt.plot(res_dict[l][:,1], res_dict[l][:,0], marker = 'o', label = "L="+str(l))
    
plt.xlabel("Temperature T")
plt.ylabel("$ \\xi $ / L")
plt.legend()
```
